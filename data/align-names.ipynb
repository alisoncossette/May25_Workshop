```python
import pandas as pd

holdings_df = pd.read_csv('Asset_Manager_Holdings.csv')
#holdings_df
```

```python
companies_df = pd.read_csv('Company_Filings.csv')
```

```python
companies_df['name'] = companies_df['name'].str.strip().str.upper()
companies_df['form10KUrls'] = companies_df['form10KUrls'].str.replace('\\', '/')
companies_df.to_csv('Companies.csv', index=False)
companies_df
```

```python
aligned_holdings_df = holdings_df.merge(companies_df[['cusip', 'name']], on='cusip', how="left")
aligned_holdings_df['companyName'] = aligned_holdings_df['name'].fillna(aligned_holdings_df['companyName'])
aligned_holdings_df.drop(columns=['name'], inplace=True)
aligned_holdings_df.to_csv('Asset_Holdings.csv', index=False)
aligned_holdings_df
```
